#**Rangkuman Matplotlib**

`Matplotlib` adalah pustaka plotting yang hebat dalam Python yang digunakan untuk membuat visualisasi statistik, animasi, dan interaktif. Tujuan utama Matplotlib adalah menyediakan alat dan fungsi bagi pengguna untuk merepresentasikan data secara grafis, sehingga lebih mudah untuk dianalisa dan dipahami. #Box Plot

`Keterangan`\
Box plot, juga dikenal sebagai diagram kotak atau box and whisker plot, adalah alat visualisasi statistik yang digunakan untuk menggambarkan distribusi data numerik. Grafik ini menyajikan ringkasan lima angka dari data, yaitu nilai minimum, kuartil pertama (Q1), median (Q2), kuartil ketiga (Q3), dan nilai maksimum. Box plot memberikan gambaran yang jelas tentang sebaran data, serta membantu dalam mengidentifikasi outlier atau nilai ekstrem.\
`Komponen Box Plot`
- Kotak (Box): Mewakili rentang interkuartil (IQR), yaitu selisih antara kuartil pertama dan kuartil ketiga (Q3 - Q1). Kotak ini menunjukkan 50% dari data yang terletak di antara Q1 dan Q3.
- Garis Median: Garis di dalam kotak yang menunjukkan nilai median dari dataset.
- Whiskers: Garis yang menjulur dari kotak ke arah nilai maksimum dan minimum yang tidak dianggap sebagai outlier. Panjang whiskers biasanya ditentukan hingga 1.5 kali IQR dari Q1 dan Q3.
- Outlier: Titik data yang berada di luar whiskers, sering kali dianggap sebagai nilai ekstrim.\
`Kegunaannya`
- Visualisasi Distribusi Data
- Identifikasi Outlier
- Perbandingan Antara Kelompok
- Analisis Variabilitas
- Membantu Dalam Analisis Data Eksploratif
- Penyajian Data Yang Ringkas


#Berikut adalah gambaran Box Plot


```python
data = np.random.randn(1000)
plt.boxplot(data)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Boxplot')
plt.show()
```

#Linspace

`Keterangan`\
Linspace adalah fungsi dalam library NumPy di Python yang digunakan untuk menghasilkan array satu dimensi yang berisi angka-angka dengan jarak yang sama dalam interval tertentu. Fungsi ini sangat berguna ketika Anda ingin membagi rentang nilai menjadi sejumlah titik yang terdistribusi secara merata. Linspace menerima parameter untuk menentukan nilai awal, nilai akhir, dan jumlah titik yang ingin dihasilkan.\
`Sintaks Linspace`

numpy.linspace(start, stop, num=50, endpoint=True)\
atau\
numpy.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None, axis=0)\

- start: Nilai awal dari interval.
- stop: Nilai akhir dari interval (jika endpoint=True, nilai ini termasuk dalam hasil).
- num: Jumlah total titik yang ingin dihasilkan (default adalah 50).
- endpoint: Jika True, nilai akhir akan disertakan dalam hasil.\
`Kegunaannya`
- Pembuatan Data Untuk Grafik
- Simulasi Dan Analisis Numerik
- Pembagian Rentang Angka
- Pengujian Fungsi
- Visualisasi Data


# Berikut adalah gambaran Linspace


```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Plot')
plt.show()
```

#Box Plot

`Keterangan` \
Box plot, juga dikenal sebagai 'box and whisker plot,' adalah jenis grafik visualisasi data yang digunakan untuk menunjukkan distribusi data. Grafik ini memberikan gambaran visual tentang rentang data, termasuk nilai minimum, kuartil pertama (Q1), median (50%), kuartil ketiga (Q3), dan nilai maksimum. Selain itu, box plot juga menonjolkan 'whiskers' yang mengindikasikan variasi di luar kuartil atas dan bawah, serta menunjukkan adanya outlier atau data yang jauh berbeda dari lainnya2 \
`Komponen Box Plot`
- Value Minimum: Nilai terkecil dalam data.
- Kuartil Pertama (Q1): Median dari separuh pertama data.
- Median (Q2): Rata-rata dari nilai ke-5 dan ke-6 dalam urutan ascending.
- Kuartil Ketiga (Q3): Median dari separuh kedua data.
- Value Maximum: Nilai terbesar dalam data.
- Whiskers: Garis-garis yang menjalar dari kotak menuju nilai minimum dan maksimum, menunjukkan variasi di luar kuartil.
- Outlier: Titik-titik yang berada di luar whiskers, sering dianggap sebagai nilai ekstrem. \
`Kegunaannya`
- Mengidentifikasi Rentang Data
- Memvisualisasikan Kuartil
- Mendeteksi Outlier
- Analisis Data Eksploratif
- Membandingkan Distribusi Antar Kelompok
- Studi Anomali
- Analisis Multivariat
- Data Tidak Normal

#Gambaran

```python
tips = sns.load_dataset("tips")
ax = sns.boxplot(x="day", y="total_bill", data=tips)
plt.title('boxplot')
plt.show()
```

#Displot

`Keterangan` \
Displot dapat digunakan untuk membuat histogram, yang mengelompokkan data ke dalam interval (atau bins) dan menghitung frekuensi data dalam setiap interval tersebut. Selain histogram, displot juga dapat menampilkan Kernel Density Estimate (KDE), yang memberikan gambaran halus tentang distribusi data. Dengan menggunakan parameter seperti `bins`, `hist`, dan `kde`, pengguna dapat menyesuaikan tampilan dan informasi yang ditampilkan. \
`Kegunaannya`
- Visualisasi Distribusi Data
- Identifikasi Outlier
- Perbandingan Distribusi
- Analisis Tren

```python
tips = sns.load_dataset('tips')
sns.displot(tips['tip'], bins=20, kde=True)
plt.show()
```
